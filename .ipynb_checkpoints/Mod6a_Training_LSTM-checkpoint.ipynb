```python
Y_train_reshaped = TrainY[0:2000,1]
Y_val = TrainY[2000:2341,1]
```

```python
X_train_final = X_train_reshaped[0:2000,:,:]
```

```python
print(X_train_final.shape)
```

```python
print(X_val.shape)
```

```python
from tensorflow.keras import layers
from tensorflow.keras import models
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from tensorflow.keras import optimizers
```

```python
from keras import backend as K

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision
```

```python
from tensorflow import keras

opt = keras.optimizers.Adam(learning_rate=0.01)
```

```python
model = Sequential()
model.add(LSTM(8, activation='relu', return_sequences = True, input_shape=(1, 12)))
model.add(layers.Dropout(0.25))
model.add(LSTM(8, activation='relu', return_sequences = True, input_shape=(1, 12)))
model.add(layers.Dropout(0.25))
model.add(LSTM(8, activation='relu', input_shape=(1, 12)))
model.add(layers.Dropout(0.25))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer="adam", metrics=['acc'])
model.summary()
```

```python
history = model.fit(X_train_final, Y_train_reshaped, batch_size=64, epochs=4000,
          validation_data=(X_val, Y_val), verbose=2)
```

```python
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```
