```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import asarray
from numpy import savetxt
```

### Carregando os dados

```python
data = pd.read_csv("data/trabalho_final4.csv", parse_dates=True)
data.index = data["date"]
data.index = pd.to_datetime(data.index)
data.index.name = "date"
data.drop(["date"], inplace = True, axis = 1)
```

```python
data.isna().sum()
```

```python
data.describe()
```

### Separando set de treino e set de teste

```python
TrainLen = int(len(data) * 0.80)
TestLen = len(data) - TrainLen
print("Tamanho do set de treino:", TrainLen, "\nTamanho do set de teste:", TestLen)
```

```python
TrainData = data.iloc[0:TrainLen,:] 
TestData  = data.iloc[TrainLen:len(data),:]
print("Formato do set de treino:", TrainData.shape,"\nFormato do set de treino:", TestData.shape)
```
